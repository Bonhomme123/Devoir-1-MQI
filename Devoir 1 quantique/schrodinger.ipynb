```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-0.3, 3, 1000)
a = 4
Psi = 2*np.sqrt(a**3)*x*np.exp(-a*x)
Psi[x<0] = 0
fig = plt.figure()
plt.plot(x, Psi, 'k')
plt.xlabel('x')
plt.ylabel("$\psi$")
fig.savefig("graph1.pdf")
```
